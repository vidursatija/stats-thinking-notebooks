```python
from nhanes.load import load_NHANES_data, load_NHANES_metadata
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nhanes_data = load_NHANES_data()
```

```python
nhanes_data.keys()
```

```python
smoking_age = nhanes_data[['AgeStartedSmokingCigarettesRegularly']]
smoking_age.dropna(inplace=True)
smoking_age = smoking_age[smoking_age['AgeStartedSmokingCigarettesRegularly'] < 100]
plt.hist(smoking_age['AgeStartedSmokingCigarettesRegularly'].values, bins=20); plt.show()
```

```python
sample50_50 = np.random.choice(smoking_age['AgeStartedSmokingCigarettesRegularly'].values, size=(100, 100))
sample50_50_mean = np.mean(sample50_50, axis=-1)
plt.hist(sample50_50_mean); plt.show()
```

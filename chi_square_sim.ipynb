```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
```

```python
RUNS = 100000
df = 20
vals = np.zeros(RUNS)
for run in range(RUNS):
    set_vals = np.random.normal(size=df)
    set_vals = np.power(set_vals, 2)
    vals[run] = set_vals.sum()

print(vals.mean())
plt.hist(vals, bins=100)
plt.show()
```
